```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = r'C:\Windows\Fonts\HANDotum.ttf'
font = font_manager.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

f = open('유무임승하차.csv', 'r', encoding='cp949')
data= csv.reader(f)
next(data)
sub_ride ={}
for row in data:
    if row[1] not in sub_ride:
        sub_ride[row[1]]=[]
        sub_ride[row[1]].append(int(row[4]))
        sub_ride[row[1]].append(int(row[5]))
        sub_ride[row[1]].append(int(row[6]))
        sub_ride[row[1]].append(int(row[7]))
    else:
        sub_ride[row[1]][0] += int(row[4])
        sub_ride[row[1]][1] += int(row[5])
        sub_ride[row[1]][0] += int(row[6])
        sub_ride[row[1]][1] += int(row[7])

# for key, value in sub_ride.items():
#     print('{} : {}'.format(key, value))

f, axes = plt.subplots(9,3)
f.set_size_inches((20, 50))
plt.subplots_adjust(wspace=0.1, hspace=0.1)
cnt = 0
for key, value in sub_ride.items():
    i = cnt//3
    j = cnt%3
    axes[i, j].pie(value, labels=['유임승차', '유임하차', '무임승차', '무임하차'], autopct='%.1f%%')
    axes[i, j].set_title(key, fontsize=16)
    cnt+=1
plt.show()
```
