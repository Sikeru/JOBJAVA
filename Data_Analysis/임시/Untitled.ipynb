```python
x=np.random.randint(-100,100,1000)
y=np.random.randint(-100,100,1000)
#size = np.random.randint(100)

mask1 = abs(x)>50
mask2 = abs(y)>50

x=x[mask1+mask2]
y=y[mask1+mask2]

size = np.random.randint(0, 100,len(x))
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
