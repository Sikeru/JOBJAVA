```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = r'C:\Windows\Fonts\HANDotum.ttf'
font = font_manager.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

with open('ta_all.csv', 'r', encoding='cp949') as f:
    data = csv.reader(f, delimiter=',')
    #print(data)
    header = next(data)

    aug=[]
    jan=[]

    for row in data:
        if row[0][5:7] =='01':
            jan.append(float(row[4]))
        if row[0][5:7] =='08':
            aug.append(float(row[4]))

plt.figure(figsize=(20, 8))
plt.title('대전의 1월과 8월의 최고기온 빈도수 그래프')
plt.hist(jan, bins=100, color='r', label='1월')
plt.hist(aug, bins=100, color='g', label='8월')
plt.show()

plt.boxplot(jan)
plt.boxplot(aug)
plt.show()
```
