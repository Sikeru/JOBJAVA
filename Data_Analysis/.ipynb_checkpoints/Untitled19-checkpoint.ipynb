```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = r'C:\Windows\Fonts\HANDotum.ttf'
font = font_manager.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

f = open('시간대별.csv', 'r', encoding='cp949')
data= csv.reader(f)
header = next(data)
next(data)

# time_ride = {}
# time_ride[0]=[]
# time_ride[0].append('승차')
# time_ride[0].append('하차')
# print(time_ride)

time_ride = {}
for row in data:
    for i in range(10,14, 2):
        if row[3] not in time_ride:
            time_ride[row[3]]=[]
            time_ride[row[3]].append(int(row[i]))
            time_ride[row[3]].append(int(row[i+1]))
        else:
            #print(time_ride[row[3]])
            time_ride[row[3]][0]+=int(row[i])
            time_ride[row[3]][1]+=int(row[i+1])

print(time_ride)

        
        
```
