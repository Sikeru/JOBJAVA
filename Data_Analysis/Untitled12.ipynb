```python
import random
x=[]
y=[]
s=[]
for i in range(100):
    x.append(random.randint(1, 100))
    y.append(random.randint(1, 100))
    s.append(random.randint(1, 200))
plt.scatter(x, y, s=s, c=range(100), cmap='jet', alpha = 0.8)
plt.colorbar()
plt.show()
```
