```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = r'C:\Windows\Fonts\HANDotum.ttf'
font = font_manager.FontProperties(fname=font_path, size=16).get_name()
rc('font', family=font)

f = open('교통카드.csv', 'r', encoding='cp949')
data= csv.reader(f)
next(data)

sub_ride = {}
for row in data:
    if row[1] not in sub_ride:
        sub_ride[row[1]] = int(row[4])
    else:
        sub_ride[row[1]] +=int(row[4])

station = []
ride_num = []
for key, value in sub_ride.items():
    station.append(key)
    ride_num.append(value)
    
plt.figure(figsize=(30,10))
plt.bar(station, ride_num)
plt.show()
```
