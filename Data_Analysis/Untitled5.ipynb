```python
import matplotlib.pyplot as plt
import csv

f = open('ta_all.csv')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
    
plt.boxplot(month)
plt.show()
```
